# Обробка властивості Name

```python
all_data['name_title'] = all_data.Name.apply(lambda x: x.split(',')[1].split('.')[0].strip())
#mr., ms., master. 
```

```python
titanic_train['name_title'] = titanic_train.Name.apply(lambda x: x.split(',')[1].split('.')[0].strip())
titanic_test['name_title'] = titanic_test.Name.apply(lambda x: x.split(',')[1].split('.')[0].strip())
titanic_train['name_title'].value_counts()
```

```python
all_data.Pclass = all_data.Pclass.astype(str)
titanic_train.Pclass = titanic_train.Pclass.astype(str)
titanic_test.Pclass = titanic_test.Pclass.astype(str)
```

```python
titanic_train.columns
```

```python
all_dummies = pd.get_dummies(all_data[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked','name_title']])
train_dummies = pd.get_dummies(titanic_train[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked','name_title']])
test_dummies = pd.get_dummies(titanic_test[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked','name_title']])
```

```python
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
plt.figure(figsize=(12, 12))
X = train_dummies.astype(float)
X.head()
```

```python
scaler = StandardScaler()
train_scaled = scaler.fit_transform(X)

pca = PCA(n_components=29)
pca_representation = pca.fit_transform(train_scaled)

plt.scatter(pca_representation[:, 0], pca_representation[:, 1])
plt.figure(figsize=(12, 12))
```

```python
X = pca_representation.astype(float)

y_pred = KMeans(max_iter=50, n_clusters=2, random_state=None).fit_predict(X)

plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
```

```python
y_pred = KMeans(max_iter=500, n_clusters=2, random_state=None).fit_predict(X)

plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
```

```python
y_pred = DBSCAN(eps=3, min_samples=2).fit_predict(X)

plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
```

```python
y_pred = DBSCAN(eps=30, min_samples=2).fit_predict(X)

plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
```

```python
y_pred = DBSCAN(eps=5, min_samples=5).fit_predict(X)

plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_pred,cmap='rainbow')
```
